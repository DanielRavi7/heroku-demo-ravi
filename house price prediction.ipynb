```python
x=df.drop(columns=['HOUSEPRICE','MAINTENANCE (MONTHLY)','TOTAL FLOORS'])
y=df['HOUSEPRICE']
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest =  train_test_split(x,y,test_size=0.3,random_state=102)
```

```python
catvar=df.select_dtypes(include=['object']).columns
catvar
```

```python

catego=['TYPE', 'FURNISHING', 'CONSTRUCTION STATUS', 'LISTED BY', 'FACING',
       'LOCALITY', 'STATE']
from catboost import CatBoostRegressor

catboo = CatBoostRegressor(iterations=220,
                            random_state = 2021, od_type = 'Iter',
                            eval_metric="R2",learning_rate=0.09,depth=2,l2_leaf_reg=5,bagging_temperature=2
                            ,border_count=255,grow_policy='Lossguide',max_leaves=140)
catboo.fit(xtrain, ytrain,cat_features=catego,eval_set=(xtest, ytest),plot=True)
```

```python

from sklearn.metrics import r2_score
x_pred = catboo.predict(xtrain)
print(r2_score(ytrain,x_pred))


# In[44]:


y_pred = catboo.predict(xtest)
r2_score(ytest,y_pred)
```
